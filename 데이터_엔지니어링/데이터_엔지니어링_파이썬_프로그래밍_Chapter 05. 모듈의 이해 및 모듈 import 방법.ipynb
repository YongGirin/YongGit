#### import 
 - import를 사용하여 해당 모듈 전체를 import

```python
import math 
```

```python
math.pi
```

```python
math.cos(100)
```

#### from import 
 - 해당 모듈에서 특정한 타입만 import

```python
from math import pi
```

```python
from math import cos
```

```python
pi
```

```python
cos(100)
```

#### \* 임포트
 - 해당 모듈내에 정의된 모든 것을 import
 - 일반적으로 사용이 권장되지 않음

```python
sin(100)
# 현재 sin함수가 정의되어있지 않음
```

```python
from math import * # 모든 기능 불러옴
# 함수 이름이 중복될 수 있기 때문에 테스트 용도 이외에는 사용하지 않는것을 권장
```

```python
sin(100)
```

```python
e
```
